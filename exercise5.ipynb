```python
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
def clip_points_to_polygon(shapefile, points):
    boundary = shapefile['geometry']
    clipped_points = gpd.clip(points, boundary)
    bounds = boundary.plot(color="white", edgecolor='black')
    populated_places = clipped_points.plot(marker='o', color="orange")
    plt.show()
    return
```

```python
# Read in shapefile, reproject, separate Alabama
states = gpd.read_file("C:/Users/cclark/Documents/408/Class_Exercise_5/tl_2025_us_state/tl_2025_us_state.shp")
states.to_crs(5070)
alabama = states[states["STUSPS"] == "AL"]

# Read in populated places file (points), reproject
points = gpd.read_file("C:/Users/cclark/Documents/408/Class_Exercise_5/ne_10m_populated_places/ne_10m_populated_places.shp")
points.to_crs(5070)
```

```python
clip_points_to_polygon(alabama, points)
```
