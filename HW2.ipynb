```python
import geopandas as gpd
import numpy as np
```

```python
alabama_shp = gpd.read_file("C:/Users/Craft/AES408/Data/cb_2024_01_place_500k/cb_2024_01_place_500k.shp")
```

```python
# Reproject to NAD83 projection for Alabama (meters)
alabama_shp = alabama_shp.to_crs(epsg=5070)

# Define perimeter for each town
alabama_shp['PERIMETER'] = alabama_shp.geometry.length

# Define a circularity index for each town; define threshold
alabama_shp['CIRCULARITY'] = (4 * np.pi * alabama_shp['ALAND']) / (alabama_shp['PERIMETER']**2)
threshold = 0.8

circular_towns = alabama_shp[alabama_shp['CIRCULARITY'] >= threshold]
print(circular_towns['NAME'])
```
